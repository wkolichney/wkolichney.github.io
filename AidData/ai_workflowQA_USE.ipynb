````python
import requests
from bs4 import BeautifulSoup
from llama_cpp import Llama
import pandas as pd
import json
import time

# ✅ Load AI Model (Llama 3, Mistral-7B, or another model)
model_path = 'C:/Users/wikku/.llama/checkpoints/Llama-Guard-3-8B/llama-pro-8b.Q5_K_M.gguf'  
llm = Llama(model_path=model_path, n_ctx=1024, n_batch=512, use_mmap=False)

# ✅ Function to Scrape Webpage Text
def fetch_web_text(url):
    try:
        response = requests.get(url, timeout=10, headers={"User-Agent": "Mozilla/5.0"})
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            return soup.get_text()
        else:
            print(f"❌ Failed to fetch {url} - Status Code: {response.status_code}")
            return None
    except Exception as e:
        print(f"❌ Error fetching {url}: {e}")
        return None

# ✅ Load Your Dataset
df = pd.read_csv("C:/Users/wikku/cleaner_internal_docp_ddc.csv")  # Update with your actual dataset filename

# ✅ Store AI decisions
results = []

# ✅ Iterate Through Each Loan Entry in the Dataset
for index, row in df.iterrows():
    url = row["URL"]
    
    # 🔹 Skip rows without a valid URL
    if not isinstance(url, str) or not url.startswith("http"):
        continue  

    # 🔹 Scrape Loan Webpage
    web_text = fetch_web_text(url)
    if not web_text:
        continue

    # 🔹 Structure the AI Prompt
    prompt = f"""
    You are an AI financial analyst reviewing loan data.

    ## **Dataset Entry**
    - **AidData Record ID:** {row['AidData Record ID']}
    - **Debt Data Type (dataset label):** {row['Debt Data Type']}
    - **Date:** {row['Date']}
    - **Observed Amount:** {row['Observed Amount']}
    - **Observed Amount Currency:** {row['Observed Amount Currency']}
    - **Recipient:** {row['Recipient']}
    - **Title:** {row['Title']}
    - **Description:** {row['Description']}
    - **Commitment Date (if available):** {row.get('Commitment Date (MM/DD/YYYY)', 'Unknown')}
    - **Interest Rate Type:** {row.get('Interest Rate Type', 'Unknown')}
    - **Source Type:** {row['Source Type']}
    - **URL (Loan Source):** {row['URL']}

    ## **Extracted Web Content**
    ```
    {web_text}
    ```

    ### **Task**
    1️⃣ **Compare the dataset's observed amount and commitment date to the text in the source URL.**  
    2️⃣ **Determine if the dataset's observed amount is truly a loan-level figure (e.g., amount outstanding) or if it mistakenly records a commitment amount.**  
    3️⃣ **If possible, extract the correct loan-level data from the source.**  
    4️⃣ **Return a decision in JSON format:**
    ```json
    {{
      "decision": "Valid Loan-Level Data / Misclassified Commitment Amount / Uncertain",
      "reasoning": "Explain why based on the extracted source text.",
      "corrected_amount_if_available": "Extracted loan-level amount if different",
      "commitment_amount_if_available": "Extracted commitment amount if different"
    }}
    ```
    """

    # 🔹 Send Prompt to AI
    response = llm(prompt)
    
    # 🔹 Parse AI Response (Ensure JSON Format)
    try:
        ai_output = json.loads(response["choices"][0]["text"])
        ai_output["AidData Record ID"] = row["AidData Record ID"]
        ai_output["URL"] = row["URL"]
        results.append(ai_output)
    except json.JSONDecodeError:
        print(f"⚠ Error parsing AI response for AidData Record ID {row['AidData Record ID']}")

    # 🔹 Respect API Rate Limits
    time.sleep(1)  # Avoid too many requests at once

# ✅ Convert Results into a DataFrame
results_df = pd.DataFrame(results)

# ✅ Save AI Validation Results
results_df.to_csv("loan_ai_validation_results.csv", index=False)

print("✅ AI Validation Completed! Results saved to 'loan_ai_validation_results.csv'")
````
